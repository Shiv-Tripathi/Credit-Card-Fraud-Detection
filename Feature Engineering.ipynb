# F_class_if for Feature Selection

```python
X = df.iloc[:,0:30]
Y = df.iloc[:,-1]
from sklearn.feature_selection import f_classif
feature_set_1 = f_classif(X,Y)
print(feature_set_1)
```

```python
dff = pd.DataFrame(data=feature_set_1, index=["The set of F values", "The set of p-values"])
```

```python
feature = {"Features" : ['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']}
dff = dff.transpose()
dff["Features"] = pd.DataFrame(data = feature)
dff
```

```python
ff = dff.sort_values('The set of F values', ascending=True).head(20)
feature_list = ff["Features"]
X.drop(feature_list, axis=1, inplace = True)
X
```
